```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
data = pd.read_csv (r'1.01.+Simple+linear+regression.csv')
```

```python
data
```

```python
data.describe()
```

```python
x1= data['SAT']
Y = data ['GPA']
```

```python
plt.scatter(x1, Y)
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```

```python
x=sm.add_constant(x1)
results = sm.OLS(Y,x). fit()
results.summary()
```

```python
plt.scatter(x1, Y)
yhat = 0.275 + 0.0017*x1
line = plt.plot(x1,yhat, lw=2, c='red', label='regression line')
plt.xlabel('SAT SCORES', fontsize = 15)
plt.ylabel('GPA SCORES', fontsize = 20)
plt.show()
```
